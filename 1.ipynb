```python
import requests

headers = {
    "User-Agent": "FinSwarm Research lluongddung1@gmail.com"
}

url = "https://www.sec.gov/files/company_tickers.json"
data = requests.get(url, headers=headers).json()

def get_cik_by_ticker(ticker):
    ticker = ticker.upper()
    for _, item in data.items():
        if item["ticker"].upper() == ticker:
            return str(item["cik_str"]).zfill(10), item["title"]
    return None, None

for ticker in ["TSLA", "AAPL", "MSFT", "AMZN", "NVDA"]:
    cik, name = get_cik_by_ticker(ticker)
    print(ticker, cik, name)
```
